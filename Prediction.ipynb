WE WILL USE STEMMING AS WELL TO REPLACE [LOVED , LOVER , LOVE] INTO [LOVE , LOVE , LOVE]

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem('loved') , ps.stem('loving'),ps.stem('love')
```

```python

def stem(x):
    y = []
    for i in x.split(' '):
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: stem(x))
```

```python
new_df
```

TEXT TO VECTORS CONVERSIONS

stop words wouldn't be considered in a vector

4806 movies and 5000 vectors point for each of them 

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()[0:100]
```

```python
ps.stem('actual')
```

USE COSINE TO CALCULATE THE DISTANCE

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

(4806, 4806) for each movie it has find the distanbce

```python
similarity
```

sorted(list(enumerate(similarity[new_df[new_df['title']=='Avatar'].index[0]])),reverse=True,key=lambda x :x[1])[1:6]

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distance = similarity[movie_index]
    movie_list  = sorted(list((enumerate(distance))),reverse=True,key= lambda x :x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[[i[0]]].title.to_numpy()[0])
```

```python
recommend('Batman Begins')
```
